```python
# import the necessary packages
import numpy as np
import glob
import cv2
 
def auto_canny(image, sigma=0.33):
    # compute the median of the single channel pixel intensities
    v = np.median(image)
    # apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v)) 
    upper = int(min(255, (1.0 + sigma) * v))
    #print(lower)
    #print(upper)
    edged = cv2.Canny(image, lower, upper)
    edged1 = cv2.Canny(image, 200, 400)
 
    # return the edged image
    return edged,edged1
orignal = cv2.imread('datare/9.jpg')
cannyed , cannyed1 = auto_canny(orignal)
cv2.imshow('original',orignal)
cv2.imshow('Otsu threshold',cannyed)
cv2.imshow('test threshold',cannyed1)
cv2.imwrite("test1.jpg",cannyed1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy
def avg(file):
    myimg = cv2.imread(file)
    #cv2.imshow('t',myimg)
    #a = numpy.average(myimg)
    avg_color_per_row = numpy.average(myimg, axis=0)
    avg_color = numpy.average(avg_color_per_row, axis=0)
    #print(avg_color)
    return avg_color
import os
files = os.listdir('datare')
lis = []
i=1
for file in files:
    av = avg('datare/'+file)
    lis.append(av)
    
print(lis)
```

```python
from PIL import Image
import os
def avge(path):
    im = Image.open(path)
    im_grey = im.convert('LA') # convert to grayscale
    #im_grey = im
    width, height = im.size

    total = 0
    for i in range(0, width):
        for j in range(0, height):
            total += im_grey.getpixel((i,j))[0]

    mean = total / (width * height)
    #print(mean)
    return mean
path = 'cannyout'
files = os.listdir(path)
lis = []
i=1
for file in files:
    av = avge('cannyout/'+file)
    lis.append(av)
    
print(lis)
```
